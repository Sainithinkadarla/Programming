### Pip commands

    pip install <package.

    pip list -o → outdated packaages

    pip list →  list of packages

    pip search <packge> → search for a package

    pip uninstall <package>  → uninstall a package

    pip install -U <package> → to update a packagenen

    pip freeze → to get list of previously installed packages

    pip freeze > <filename> → to get list of packages in file

    pip install -r <filename> → to install thepackages in filename

    pip freeze —local → to get the list of packages installed in the current working python venv

### Virtual Environment

    pip install virtualenv

    virtualenv <env_name>  → to create a virtual environment

    source <env_name>/bin/activate → to activate the virtual environment

    which python → to get python path 

    which pip→ to get pip path 

    deactivate → to deactivate venv

    rm -rf <venv> → to delete a virtual environment

    virtualenv -p <python path> <venv_name> → to create a venv with a specified python version


### Conda
    conda list -> To get list of available packages
    
    conda create --name <name> <list of packages with space seperated> -> To create a new environment with the given name and the packages installed in that environment    

    conda create --name <name> python=<version> -> To specify python version
    
    conda env list -> To get the list of environments in the machine

    conda remove --name <name> --all -> To remove the environments with all the packages installed
