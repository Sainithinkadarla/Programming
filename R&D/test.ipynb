```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering, Birch, DBSCAN
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn import datasets
```

```python
i = datasets.load_wine()
data = i.data
pca = PCA(n_components=2)
df = pca.fit_transform(data)
kmeans = KMeans(n_clusters=3,random_state=42)
kmeans.fit(df)
cluster_label = kmeans.labels_
centroid = kmeans.cluster_centers_
```

```python
plt.scatter(df[:,0],df[:,1],c=cluster_label,cmap='terrain',edgecolors='k',s=75)
plt.scatter(centroid[:,0],centroid[:,1],c='red',marker='X',s=120,label='centroids')
plt.legend()
plt.title('K-Means on wine dataset')
plt.xlabel('Prinipal component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
d, true_labels = make_blobs(n_samples=50, centers=4, random_state=22)

# Hierarchical clustering
linkage_matrix = linkage(d, method='complete')

# Dendrogram plot
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Points')  # Rotate the x-axis labels
plt.ylabel('Distance')

plt.show()
```

```python
data, true_labels = make_blobs(n_samples=300, centers=4, random_state=22)

# Agglomerative clustering
agg_cluster = AgglomerativeClustering(n_clusters=4)
agg_labels = agg_cluster.fit_predict(data)

# Plotting Agglomerative clustering results
plt.figure(figsize=(12, 5))

# Scatter plot
plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1], c=agg_labels, cmap='viridis', edgecolors='k')
plt.title('Agglomerative Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

```python
birch_cluster = Birch(threshold=1.5, n_clusters=4)
birch_labels = birch_cluster.fit_predict(data)

# Plotting BIRCH clustering results
plt.scatter(data[:, 0], data[:, 1], c=birch_labels, cmap='viridis', edgecolors='k')
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(data)

plt.scatter(data[:, 0], data[:, 1], c=dbscan_labels, cmap='viridis', edgecolors='k')
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
