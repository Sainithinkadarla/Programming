```python
import numpy as np
from numpy import where
from numpy import unique
from matplotlib import pyplot
from sklearn.datasets import make_classification
from sklearn.cluster import KMeans
x,_=make_classification(n_samples=1000,n_features=2,n_informative=2,n_redundant=0,n_clusters_per_class=1,random_state=4) #define model
model=KMeans(n_clusters=2)
#fit the model
model.fit(x)
#assign the cluster
yhat=model.predict(x)
#for unique clusters
clusters=unique(yhat)
#creating scatter plot for samples for each clusters
for cluster in clusters:
    #get row indexes for samples
    row=where(yhat==cluster)
    pyplot.scatter(x[row,0],x[row,1]) 
pyplot.show()
```
