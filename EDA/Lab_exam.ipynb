```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.get_dataset_names()
```

```python
data = sns.load_dataset('iris')
data
```

```python
sns.lineplot(x='sepal_length', y='petal_length', data=data)
```

```python
diamonds = sns.load_dataset('diamonds')

sns.distplot(diamonds['price'],hist=False, color ='green', vertical= True)
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=data, kind = 'reg')
sns.jointplot
data
```

```python
sns.boxplot(y='sepal_length', x= 'species', data=data, linewidth=2, palette='Set1')
```
