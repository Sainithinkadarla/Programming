# Covid data set

#### Importing the data set

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset="Latest Covid-19 India Status.csv"

csvdata = pd.read_csv(dataset)

csvdata
```

### Viewing the columns

```python
index_columns= csvdata.columns
print(index_columns,end="\n\n\n")
print(csvdata.info())
```

#### Filtering the deaths with respective to top 10 states



```python
deaths_filtered= csvdata.filter(['State/UTs','Deaths'])
sorted_deaths = deaths_filtered.sort_values('Deaths') 
min_deaths = (deaths_filtered.sort_values('Deaths', ascending=True)).head(10)
max_deaths = (deaths_filtered.sort_values('Deaths', ascending=False)).head(10)
```

#### Plotting the data 

```python
min_deaths.plot.barh('State/UTs','Deaths')
max_deaths.plot.barh('State/UTs','Deaths')
```

### Filtering the data with respective to active cases


```python
active_cases_filtered = csvdata.filter(['State/UTs', 'Active'])
min_AC = (active_cases_filtered.sort_values('Active',ascending=True)).head(10)
max_AC = (active_cases_filtered.sort_values('Active', ascending=False).head(10))
```

```python
max_AC.plot.barh('State/UTs', 'Active')
min_AC.plot.barh('State/UTs', 'Active')
```

```python
discharged_filtered = csvdata.filter(['State/UTs', 'Discharged'])
max_D = (discharged_filtered.sort_values('Discharged')).head(10)
```

```python
max_D.plot.barh('State/UTs', 'Discharged')
```

# IPL data set

#### importing the dataset

```python
matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')
matches.info()
```

```python
deliveries.info()
```

```python
matches.head()
```

#### Matches per season

```python
seasons = matches.groupby('season')
seasons_list = list(seasons.groups.keys())
seasons_list
```

```python
matches_per_season = []
for x in seasons_list:
    matches_per_season.append(seasons.get_group(x).count()['id'])

matches_per_season
```

```python
plt.bar(range(len(matches_per_season)), matches_per_season,tick_label = seasons_list)

for index, values in enumerate(matches_per_season):
    plt.text(index,values+1, str(values))
```

#### Most utilized venues


```python
venues = matches.groupby('venue')
venues_list = list(venues.groups.keys())

matches_in_venues ={}

for x in venues_list:
    matches_in_venues[x] = (venues.get_group(x).count()['id'])

matches_in_venues
```

```python
sorted_venues = sorted(matches_in_venues.items(), key= lambda kv: (kv[1], kv[0]))

top10_venues =sorted_venues[-10:]

venue_visited = []
no_of_times_venues_utilized = []

for i in top10_venues:
    venue_visited.append(i[0])
    no_of_times_venues_utilized.append(i[1])

```

```python
plt.barh(range(len(no_of_times_venues_utilized)), no_of_times_venues_utilized, tick_label = venue_visited)

plt.xticks(rotation ='vertical')
plt.yticks(rotation ='horizontal')
```

#### Most runs by batsman


```python
most_runs = deliveries.groupby('batsman')
batsman_list = list(most_runs.groups.keys())
```

```python
batsmans_runs = {}

for i in batsman_list:
    batsmans_runs[i] = (most_runs.get_group(i).agg({'total_runs':sum})['total_runs'])

batsmans_runs.items()
```

```python
sorted_batsman_runs = sorted(batsmans_runs.items(), key=lambda kv: (kv[1], kv[0]), reverse = True)
sorted_batsman_runs
```

```python
batsman_names =[]
runs_by_batsman =[]

for x in sorted_batsman_runs:
    batsman_names.append(x[0])
    runs_by_batsman.append(x[1])

plt.barh(range(len(runs_by_batsman[:10])), runs_by_batsman[:10], tick_label =batsman_names[:10])
plt.xticks(rotation ='vertical')
```

#### Most sixes by batsman 

```python
sixes = deliveries[deliveries['batsman_runs'] == 6]
sixes = deliveries.groupby('batsman')
sixes_by_batsman_list = list(sixes.groups.keys())
```

```python
sixes_by_batsman = {}

for i in sixes_by_batsman_list:
    sixes_by_batsman[i] = sixes.get_group(i).count()[0]

sorted_sixes_by_batsman = sorted(sixes_by_batsman.items(), key= lambda kv: (kv[1], kv[0]), reverse=True)


sixes_batsman = []
no_of_sixes = []

for i in sorted_sixes_by_batsman:
    sixes_batsman.append(i[0])
    no_of_sixes.append(i[1])
```

```python
plt.barh(range(len(no_of_sixes[:10])), no_of_sixes[:10], tick_label= sixes_batsman[:10])
plt.xticks(rotation = 'vertical')
```
