# Covid data set

#### Importing the data set

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset="Latest Covid-19 India Status.csv"

csvdata = pd.read_csv(dataset)

csvdata
```

### Viewing the columns

```python
index_columns= csvdata.columns
print(index_columns,end="\n\n\n")
print(csvdata.info())
```

### Filtering the deaths with respective to top 10 states



```python
deaths_filtered= csvdata.filter(['State/UTs','Deaths'])
sorted_deaths = deaths_filtered.sort_values('Deaths') 
min_deaths = (deaths_filtered.sort_values('Deaths', ascending=True)).head(10)
max_deaths = (deaths_filtered.sort_values('Deaths', ascending=False)).head(10)
```

#### Plotting the data 

```python
min_deaths.plot.barh('State/UTs','Deaths')
max_deaths.plot.barh('State/UTs','Deaths')
```

### Filtering the data with respective to active cases


```python
active_cases_filtered = csvdata.filter(['State/UTs', 'Active'])
min_AC = (active_cases_filtered.sort_values('Active',ascending=True)).head(10)
max_AC = (active_cases_filtered.sort_values('Active', ascending=False).head(10))
```

```python
max_AC.plot.barh('State/UTs', 'Active')
min_AC.plot.barh('State/UTs', 'Active')
```

```python
discharged_filtered = csvdata.filter(['State/UTs', 'Discharged'])
max_D = (discharged_filtered.sort_values('Discharged')).head(10)
```

```python
max_D.plot.barh('State/UTs', 'Discharged')
```
